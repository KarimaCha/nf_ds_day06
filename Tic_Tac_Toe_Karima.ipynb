```python
def print_board(board):
    for line in board:
        print(line)
    print('\n')
        
def get_diagonal(board, start_index):
    end_index = 2 if start_index == 0 else 0
    diagonal = [board[start_index][0], board[1][1], board[end_index][2]]
    return diagonal
        
def run_tic_tac_toe():

    # Part 2: Set up the game (build and print board)
    board = [[(0, 0), (1, 0), (2, 0)], [(0, 1), (1, 1), (2, 1)], [(0, 2), (1, 2), (2, 2)]]
    print_board(board)
    symbols = [' X  ', ' O  ']
    
    # Part 3: Commence the game (determine which player's turn it is)
    turn = 0
    while turn <= 8:
        player = turn % 2 + 1
        symbol = symbols[player-1]
        
        # Part 4: Accept a play (allow a player to specify where they want to play)
        position = eval(input('Player {} ( {} ), where will you play? '.format(player, symbol)))
        
        while position not in [cell for line in board for cell in line]:
            print('Invalid choice! Position out of range or already taken!')
            position = eval(input('Player {} ( {} ), where will you play? '.format(player, symbol)))
        
        board[position[1]][position[0]] = symbol
        print_board(board)
        
        # Part 5: Check for a winner
        row1_won = all(cell==symbol for cell in board[0])
        row2_won = all(cell==symbol for cell in board[1])
        row3_won = all(cell==symbol for cell in board[2])
        
        col1_won = all([line[0]==symbol for line in board])
        col2_won = all([line[1]==symbol for line in board])
        col3_won = all([line[2]==symbol for line in board])
        
        diag1_won = all([cell==symbol for cell in get_diagonal(board, 0)])
        diag2_won = all([cell==symbol for cell in get_diagonal(board, 2)])
        
        won = any([row1_won, row2_won, row3_won, col1_won, col2_won, col3_won,
                   diag1_won, diag2_won])
                   
        if won:
            print('Player {} Wins!!'.format(player))
            break
        elif turn==8:  # Part 6: Tie Game (all cells are filled)
            print('Game resulted in a tie...like usual.')
            break
        
        turn += 1
        

# Part 1: Getting started (put main block at bottom of script)
if __name__ == '__main__':
    run_tic_tac_toe()
```
